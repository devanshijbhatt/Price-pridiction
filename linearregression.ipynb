```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train_set= pd.read_csv('train.csv')  
test_set = pd.read_csv('test.csv')
```

```python
X_train = train_set[['rm']]
# print(X_train)
y_train = train_set['medv']
# print(y_train)
```

```python
X_test = test_set[['rm']]
# y_test = test_set['medv']
print(X_test)
```

```python

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)  # Predictions on the training data
y_pred_test = model.predict(X_test)    # Predictions on the testing data
```

```python
# Calculate mean squared error and R-squared value on training data
mse_train = mean_squared_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

print(f'Training Data Mean Squared Error: {mse_train}')
print(f'Training Data R-squared Value: {r2_train}')
```

```python
# Visualize results for the training data
plt.scatter(X_train, y_train, color='blue', label='Actual Prices')
plt.plot(X_train, y_pred_train, color='red', linewidth=2, label='Predicted Prices')
plt.xlabel('Number of Rooms (RM)')
plt.ylabel('Median House Prices (MEDV)')
plt.title('Linear Regression: Number of Rooms vs. Median House Prices')
plt.legend()
plt.show()
```

```python
test_set['predicted_medv'] = y_pred_test

# Save the results to a new CSV file
test_set.to_csv('test_data_with_predictions.csv', index=False)
```
